## Sampling Blur images 
The blur images are from light (gaussian blur of 1 pixel), medium (gaussian blur of 5 pixels), and heavy (gaussian blur of 10 pixels).
 We sample equally from each blur amount to create a consolidated blur dataset. 

## Training Set

### Generating Blur Visible

```python
from PIL import Image, ImageFilter
import os

input_dir = "D:/FLIR/train/RGB_adjusted"
output_dir_light = "D:/FLIR/train/visible_blur_light"
output_dir_medium = "D:/FLIR/train/visible_blur_medium"
output_dir_high = "D:/FLIR/train/visible_blur_high"


for file in os.listdir(input_dir):
    full_file = os.path.join(input_dir, file)
    img = Image.open(full_file)
    img_light = img.filter(ImageFilter.GaussianBlur(1))
    img_medium = img.filter(ImageFilter.GaussianBlur(5))
    img_high = img.filter(ImageFilter.GaussianBlur(10))
    output_file_light = os.path.join(output_dir_light, file)
    output_file_medium = os.path.join(output_dir_medium, file)
    output_file_high = os.path.join(output_dir_high, file)
    img_light.save(output_file_light)
    img_medium.save(output_file_medium)
    img_high.save(output_file_high)
    
```

# Generating Training Datasets and Labels 

## Oversampled dataset 


###  Train Set

```python
input_dir = "D:/FLIR/train/yolo_labels" 
file_list = []
no_label_list = []
def is_non_zero_file(fpath):  
    return os.path.isfile(fpath) and os.path.getsize(fpath) > 0

for f in os.listdir(input_dir): 
    path = os.path.join(input_dir, f) 
    if is_non_zero_file(path):
        file_list.append(f)
    else: 
        no_label_list.append(f)


file_list = [f.replace('.txt', '.jpg') for f in file_list]   
     


print(file_list)


print (len(file_list))


    
    
```

```python
#Oversample parameters 

from math import floor

#percentage of clean samples
clean_rate = 0.5 

#percentage of dirty samples 
dirty_rate = (1 - clean_rate) 


#percentage of multi sensor dirty samples
multi_sensor_dirty_rate = (0.1 * dirty_rate) 

#percentage of fog samples 
fog_rate = ((clean_rate - multi_sensor_dirty_rate)/3.)

#percentage of blurred samples 
blur_rate = ((clean_rate- multi_sensor_dirty_rate)/3.)

#percentage of dropout samples 
drop_rate = ((clean_rate - multi_sensor_dirty_rate)/3.)

#clean_rate *=100
#dirty_rate *=100
#multi_sensor_dirty_rate *=100
#fog_rate *=100
#blur_rate *=100
#drop_rate *=100

print (clean_rate, dirty_rate, multi_sensor_dirty_rate, fog_rate, blur_rate, drop_rate)


cumulative_multi_sensor_dirty_rate = clean_rate + multi_sensor_dirty_rate 
cumulative_fog_rate = cumulative_multi_sensor_dirty_rate + fog_rate 
cumulative_blur_rate = cumulative_fog_rate + blur_rate
cumulative_drop_rate = cumulative_blur_rate + drop_rate 
print (cumulative_drop_rate)
print (cumulative_blur_rate)

input_dir = "D:/FLIR/train/RGB_adjusted" 


import random 

clean_counter = 0
multi_sensor_counter = 0 
fog_counter = 0 
blur_counter = 0
drop_counter = 0
dirty_counter = 0 
counter = 0 

#clean directories 

clean_visible = "D:/FLIR/train/RGB_adjusted" 
clean_lwir = "D:/FLIR/train/thermal_8_bit_adjusted" 

#fog directories 
fog_visible = "D:/FLIR/train/visible_fog_sampled"
fog_lwir = clean_lwir 

#blur directories 
blur_visible = "D:/FLIR/train/visible_blur_sampled"
blur_lwir = "D:/FLIR/train/lwir_blur_sampled"

#drop directories 
drop_visible = "D:/FLIR/train/visible_dropout"
drop_lwir = "D:/FLIR/train/lwir_dropout"


#output directories 
output_visible = "D:/FLIR/train/visible_gating"
output_lwir = "D:/FLIR/train/lwir_gating"

os.mkdir(output_visible)
os.mkdir(output_lwir)

list_of_files = file_list
list_of_files+= file_list
random.shuffle(list_of_files)
print (len(list_of_files))

for f in list_of_files:
    rand = random.random()
    counter +=1
    f_lwir = f.replace(".jpg", ".jpeg")
    if rand <= clean_rate: #clean
        #print (f)
        og_lwir_path = os.path.join(clean_lwir, f_lwir)
        og_visible_path = os.path.join(clean_visible, f)
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
        
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
    
        clean_counter+=1
    elif rand > clean_rate and rand <= cumulative_multi_sensor_dirty_rate: #multisensor dirty 
        #print(f)
        rand_vis = random.randint(0,3)
        rand_lwir = random.randint(0,2)
        
        if rand_vis == 0: #fog
            og_visible_path = os.path.join(fog_visible, f)
        elif rand_vis == 1: #blur 
            og_visible_path = os.path.join(blur_visible, f)
        elif rand_vis == 2: #dropout 
            og_visible_path = os.path.join(drop_visible, f)
    
    
        if rand_lwir == 0: #blur 
            og_lwir_path = os.path.join(blur_lwir, f_lwir)
        elif rand_lwir == 1: #dropout 
            og_lwir_path = os.path.join(drop_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-d."
        f = file_name + file_extension
        
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-d."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path) 
        
        dirty_counter+=1
        multi_sensor_counter+=1 
    elif rand > cumulative_multi_sensor_dirty_rate and rand <= cumulative_fog_rate: #fog only rate 
        #print (f)
        og_visible_path = os.path.join(fog_visible, f)
        og_lwir_path = os.path.join(fog_lwir, f_lwir)
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-f."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-f."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
        dirty_counter +=1
        fog_counter +=1
        
    elif rand > cumulative_fog_rate and rand <= cumulative_blur_rate: #blur only rate 
        sensor = random.randint(0,1)
        
        if sensor == 0: #blur visible 
            og_visible_path = os.path.join(blur_visible, f)
            og_lwir_path = os.path.join(clean_lwir, f_lwir)
        elif sensor == 1: #blur lwir 
            og_visible_path = os.path.join(clean_visible, f)
            og_lwir_path = os.path.join(blur_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-b."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-b."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
    
        
        dirty_counter +=1 
        blur_counter +=1
    elif rand > cumulative_blur_rate and rand <= cumulative_drop_rate: #drop only 
        #print(f)
        sensor = random.randint(0,1)
        
        if sensor == 0: #drop visible 
            og_visible_path = os.path.join(drop_visible, f)
            og_lwir_path = os.path.join(clean_lwir, f_lwir)
        elif sensor == 1: #drop lwir 
            og_visible_path = os.path.join(clean_visible, f)
            og_lwir_path = os.path.join(drop_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-x."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-x."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
        
        dirty_counter +=1
        drop_counter +=1


print ("clean ", float(clean_counter/counter))
print ("dirty ", float(dirty_counter/counter))
print ("multi ", float(multi_sensor_counter/counter))
print ("fog ", float(fog_counter/counter))
print ("blur ", float(blur_counter/counter))
print ("drop ", float(drop_counter/counter))




```

```python
#Oversample parameters 

from math import floor

#percentage of clean samples
clean_rate = 0.95

#percentage of dirty samples 
dirty_rate = (1 - clean_rate) 


#percentage of multi sensor dirty samples
multi_sensor_dirty_rate = (0.1 * dirty_rate) 

#percentage of fog samples 
fog_rate = (dirty_rate*.3)

#percentage of blurred samples 
blur_rate = (dirty_rate*.3)

#percentage of dropout samples 
drop_rate = (dirty_rate*.3)

#clean_rate *=100
#dirty_rate *=100
#multi_sensor_dirty_rate *=100
#fog_rate *=100
#blur_rate *=100
#drop_rate *=100

print (clean_rate, dirty_rate, multi_sensor_dirty_rate, fog_rate, blur_rate, drop_rate)


cumulative_multi_sensor_dirty_rate = clean_rate + multi_sensor_dirty_rate 
cumulative_fog_rate = cumulative_multi_sensor_dirty_rate + fog_rate 
cumulative_blur_rate = cumulative_fog_rate + blur_rate
cumulative_drop_rate = cumulative_blur_rate + drop_rate 
print ("clean ", clean_rate)
print ("multisensor ", cumulative_multi_sensor_dirty_rate)
print ("fog ", cumulative_fog_rate)
print ("drop ", cumulative_drop_rate)
print ("blur ", cumulative_blur_rate)

input_dir = "D:/FLIR/val/RGB_adjusted" 


import random 

clean_counter = 0
multi_sensor_counter = 0 
fog_counter = 0 
blur_counter = 0
drop_counter = 0
dirty_counter = 0 
counter = 0 

#clean directories 

clean_visible = "D:/FLIR/val/RGB_adjusted" 
clean_lwir = "D:/FLIR/val/thermal_8_bit_adjusted" 

#fog directories 
fog_visible = "D:/FLIR/val/visible_fog_sampled"
fog_lwir = clean_lwir 

#blur directories 
blur_visible = "D:/FLIR/val/visible_blur_sampled"
blur_lwir = "D:/FLIR/val/lwir_blur_sampled"

#drop directories 
drop_visible = "D:/FLIR/val/visible_dropout"
drop_lwir = "D:/FLIR/val/lwir_dropout"


#output directories 
output_visible = "D:/FLIR/val/visible_gating_95"
output_lwir = "D:/FLIR/val/lwir_gating_95"

os.mkdir(output_visible)
os.mkdir(output_lwir)

list_of_files = file_list
#list_of_files+= file_list
random.shuffle(list_of_files)
print (len(list_of_files))

for f in list_of_files:
    rand = random.random()
    counter +=1
    f_lwir = f.replace(".jpg", ".jpeg")
    if rand <= clean_rate: #clean
        #print (f)
        og_lwir_path = os.path.join(clean_lwir, f_lwir)
        og_visible_path = os.path.join(clean_visible, f)
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
        
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
    
        clean_counter+=1
    elif rand > clean_rate and rand <= cumulative_multi_sensor_dirty_rate: #multisensor dirty 
        #print(f)
        rand_vis = random.randint(0,3)
        rand_lwir = random.randint(0,2)
        
        if rand_vis == 0: #fog
            og_visible_path = os.path.join(fog_visible, f)
        elif rand_vis == 1: #blur 
            og_visible_path = os.path.join(blur_visible, f)
        elif rand_vis == 2: #dropout 
            og_visible_path = os.path.join(drop_visible, f)
    
    
        if rand_lwir == 0: #blur 
            og_lwir_path = os.path.join(blur_lwir, f_lwir)
        elif rand_lwir == 1: #dropout 
            og_lwir_path = os.path.join(drop_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-d."
        f = file_name + file_extension
        
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-d."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path) 
        
        dirty_counter+=1
        multi_sensor_counter+=1 
    elif rand > cumulative_multi_sensor_dirty_rate and rand <= cumulative_fog_rate: #fog only rate 
        #print (f)
        og_visible_path = os.path.join(fog_visible, f)
        og_lwir_path = os.path.join(fog_lwir, f_lwir)
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-f."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-f."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
        dirty_counter +=1
        fog_counter +=1
        
    elif rand > cumulative_fog_rate and rand <= cumulative_blur_rate: #blur only rate 
        sensor = random.randint(0,1)
        
        if sensor == 0: #blur visible 
            og_visible_path = os.path.join(blur_visible, f)
            og_lwir_path = os.path.join(clean_lwir, f_lwir)
        elif sensor == 1: #blur lwir 
            og_visible_path = os.path.join(clean_visible, f)
            og_lwir_path = os.path.join(blur_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-b."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-b."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
    
        
        dirty_counter +=1 
        blur_counter +=1
    elif rand > cumulative_blur_rate and rand <= cumulative_drop_rate: #drop only 
        #print(f)
        sensor = random.randint(0,1)
        
        if sensor == 0: #drop visible 
            og_visible_path = os.path.join(drop_visible, f)
            og_lwir_path = os.path.join(clean_lwir, f_lwir)
        elif sensor == 1: #drop lwir 
            og_visible_path = os.path.join(clean_visible, f)
            og_lwir_path = os.path.join(drop_lwir, f_lwir)
            
        
        file_name = f.split(".")[0]
        file_extension = f.split(".")[1]
        file_name += "-x."
        f = file_name + file_extension
        
                
        file_name = f_lwir.split(".")[0]
        file_extension = f_lwir.split(".")[1]
        file_name += "-x."
        f_lwir = file_name + file_extension
        
        final_lwir_path = os.path.join(output_lwir, f_lwir)
        final_visible_path = os.path.join(output_visible, f)
            
        shutil.copy(og_lwir_path, final_lwir_path)
        shutil.copy(og_visible_path, final_visible_path)
        
        
        dirty_counter +=1
        drop_counter +=1


print ("clean ", clean_counter, counter, float(clean_counter/counter))
print ("dirty ", dirty_counter, counter, float(dirty_counter/counter))
print ("multi ", multi_sensor_counter, counter, float(multi_sensor_counter/counter))
print ("fog ", fog_counter, counter, float(fog_counter/counter))
print ("blur ", blur_counter, counter, float(blur_counter/counter))
print ("drop ", drop_counter, counter, float(drop_counter/counter))




```
